```python
import numpy as np
from numpy.random import normal

normal(loc=5, scale=25, size=20)
```

```python
x_axis = normal(loc=5, scale=6, size=100)
y_axis = normal(loc=3, scale=1, size=100)

x_axis2 = normal(loc=-5, scale=3, size=100)
y_axis2 = normal(loc=0, scale=3, size=100)
```

```python
%matplotlib inline
import seaborn as sns
sns.set()

ax = sns.scatterplot(x=x_axis, y=y_axis)
ax2 = sns.scatterplot(x=x_axis2, y=y_axis2)
```

```python
from sklearn.cluster import KMeans

sample_points = np.transpose(np.array([[x_axis, x_axis2],[y_axis, y_axis2]]).reshape((2,200)))

# Initialize kmeans algorithm
kmeans = KMeans(n_clusters=2)

# fit and predict labels for clusters
point_assignments = kmeans.fit_predict(sample_points)

# Visualize results
sns.scatterplot(x = sample_points[:,0], y = sample_points[:,1], hue=point_assignments)
```
